```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
import skimage 
```

```python
B5 = skimage.io.imread("F:/image processing in python/Image/Geo-Eye/pan_img.tif")
```

```python
kernel_size = 5
```

```python
#Low Pass Filters
mean_image = cv2.blur(B5,(kernel_size, kernel_size))
median_image = cv2.medianBlur(B5, kernel_size)
gaussian = cv2.GaussianBlur(B5, (kernel_size,kernel_size), 0)
```

```python
#High Pass Filters
sobel_image = cv2.Sobel(B5, 0, dx=0, dy=1, ksize= kernel_size)
laplacian_image = cv2.Laplacian(B5, -1, ksize=kernel_size)
```

```python
plt.figure(figsize=(20,20))
plt.subplot(3,2,1)
plt.imshow(B5, cmap='gray')
plt.title('Original Image')
plt.axis('off')
##Mean Filter
plt.subplot(3,2,2)
plt.imshow(mean_image, cmap='gray')
plt.title('Mean Filter')
plt.axis('off')
##Median Filter
plt.subplot(3,2,3)
plt.imshow(median_image, cmap='gray')
plt.title('Median Image')
plt.axis('off')
##Gaussian Filter
plt.subplot(3,2,4)
plt.imshow(gaussian, cmap='gray')
plt.title('Gaussian Filter')
plt.axis('off')
##Sobel Filter
plt.subplot(3,2,5)
plt.imshow(sobel_image, cmap='gray')
plt.title('Sobel Filter')
plt.axis('off')
##Laplacian Image
plt.subplot(3,2,6)
plt.imshow(laplacian_image, cmap='gray')
plt.title('Laplacian Filter')
plt.axis('off')
plt.show()
```
